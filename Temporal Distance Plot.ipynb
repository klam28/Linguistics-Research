```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
style.use("ggplot")

import matplotlib.ticker as ticker
# Convert CSV Delimit: Click Data -> Text to Column -> Tab -> Next -> Finish
```

```python
data = pd.read_csv("/Users/kerrylam/Desktop/Linguistics Research/allwords.csv")
data = pd.DataFrame(data)
```

```python
data.head()
```

```python
data.dropna(inplace = True)
# Drop Missing Data
```

```python
data.head()
```

```python
# Create Difference Column in Dataframe
# Create Variable For Previous Column 
data["Difference_Column"] = data.ix[:,"Start"]
prev_time = data["Difference_Column"].shift()
data["Start"] = (data.ix[:,"Start"] - prev_time)/1000
```

```python
data.head()
```

```python
# Create Count for x 
data.head()
n_words = range(1,len(data["Start"]) + 1)
print(n_words)
```

```python
# Plot X,Y
plt.figure(figsize=(20, 10))
plt.plot(n_words, data["Start"], c = "purple", linewidth = 1)
plt.xlabel("Word Count")
plt.ylabel("Time")
plt.title("Temporal Distance")

# Set Intervals
setaxis = plt.axes()
setaxis.xaxis.set_major_locator(ticker.MultipleLocator(10))
setaxis.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
# plt.savefig('tempgraph.png')
```
